```python
columns=["BALANCE","PURCHASES","ONEOFF_PURCHASES","CASH_ADVANCE","CREDIT_LIMIT","PAYMENTS","MINIMUM_PAYMENTS","INSTALLMENTS_PURCHASES"]

for col in columns:
    df_backup.loc[df_backup[col]<=500,col]=0
    df_backup.loc[((df_backup[col]>500) & (df_backup[col]<=1000)),col]=1
    df_backup.loc[((df_backup[col]>1000) & (df_backup[col]<=3000)),col]=2
    df_backup.loc[((df_backup[col]>3000) & (df_backup[col]<=5000)),col]=3
    df_backup.loc[((df_backup[col]>5000) & (df_backup[col]<=10000)),col]=4
    df_backup.loc[((df_backup[col]>10000)),col]=6
    
```

```python
columns=["BALANCE_FREQUENCY","PURCHASES_FREQUENCY","ONEOFF_PURCHASES_FREQUENCY","PURCHASES_INSTALLMENTS_FREQUENCY","PRC_FULL_PAYMENT","CASH_ADVANCE_FREQUENCY"]

for col in columns:
    df_backup.loc[df_backup[col]<=0.2,col]=0
    df_backup.loc[((df_backup[col]>0.2) & (df_backup[col]<=0.4)),col]=1
    df_backup.loc[((df_backup[col]>0.4) & (df_backup[col]<=0.6)),col]=2
    df_backup.loc[((df_backup[col]>0.6) & (df_backup[col]<=0.8)),col]=3
    df_backup.loc[((df_backup[col]>0.8) & (df_backup[col]<=1)),col]=4
        
    
```

```python
columns=["CASH_ADVANCE_TRX","PURCHASES_TRX"]

for col in columns:

    df_backup.loc[df_backup[col]<=10,col]=0
    df_backup.loc[((df_backup[col]>10) & (df_backup[col]<=20)),col]=1
    df_backup.loc[((df_backup[col]>20) & (df_backup[col]<=35)),col]=2
    df_backup.loc[((df_backup[col]>35) & (df_backup[col]<=60)),col]=3
    df_backup.loc[((df_backup[col]>60) & (df_backup[col]<=90)),col]=4
    df_backup.loc[((df_backup[col]>90) & (df_backup[col]<=120)),col]=5
    df_backup.loc[((df_backup[col]>120) & (df_backup[col]<=150)),col]=6
    df_backup.loc[(df_backup[col]>150),col]=7
```

```python
df_backup=df_backup.drop("CUST_ID",axis=1)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
cluster=20
cost=[]


for n in range(1,cluster):
    kmean=KMeans(n_clusters=n)
    kmean.fit(df_backup)
    cost.append(kmean.inertia_)
```

```python
fig=plt.plot(cost,marker="o",color="red")
plt.title("Elbow method")
plt.xlabel("Number of cluster")
plt.ylabel("Cost")
# for s in ["left","right","bottom","top"]:
#     fig.spines[s].set_visble(False)

plt.show()
```

```python
kmean=KMeans(n_clusters=4,init="k-means++",random_state=42)
y_kmean=kmean.fit_predict(df_backup)
labels=kmean.labels_
```

```python
silhouette_score(df_backup,kmean.labels_)
```
